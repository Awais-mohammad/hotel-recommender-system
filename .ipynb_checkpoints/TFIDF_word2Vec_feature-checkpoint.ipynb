```python
# libraries
import pandas as pd
import string
from sklearn.feature_extraction.text import TfidfVectorizer
from scipy.sparse import csc_matrix
import numpy as np
```

```python
df=pd.read_csv('.\\processed\\Sentiments.csv',low_memory=False,index_col=False)
```

```python
positiveDF=df.loc[df['Sentiments'] == 1]
negativeDF=df.loc[df['Sentiments'] == -1]
neutralDF=df.loc[df['Sentiments'] == 0]
```

```python
rTxt=df['ReviewText']
```

```python
ngramRange=(1,2)
nterms=None
```

```python
vectorizer = TfidfVectorizer(lowercase=True,ngram_range=ngramRange,min_df=5,max_df=0.8)
X = vectorizer.fit_transform(rTxt)
```

```python
unimodel=vectorizer.fit(rTxt)
```

```python
sortedVocab=[v[0] for v in sorted(vectorizer.vocabulary_.items())]
print(sortedVocab)
```

```python
ranks=np.array(np.argsort(X.sum(axis=0))).ravel()
```

```python
frequentTerms=[sortedVocab[r] for r in ranks]
finalfrequents=[f for f in frequentTerms if len(f) > 3]
```

```python
print(unimodel,ranks,len(finalfrequents))
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
import warnings
warnings.filterwarnings(action = 'ignore')
import gensim
from gensim.models import Word2Vec
```

```python
data=[]
for i in range(0,len(rTxt)):
    temp = []
    # tokenize the sentence into words
    for j in word_tokenize(rTxt[i]):
        temp.append(j.lower())
    data.append(temp)
print(data)
```

```python
# Create Skip Gram model
model2 = gensim.models.Word2Vec(data, min_count = 1,window = 5, sg = 1)
```

```python
# Print results
print('=-=-testing-=-')
model2.wv.most_similar("glass") 
```
